#  2)Average population of the cities where there are more males:

```python
#2
df1['Population'] = df1['Population'].astype(str).str.replace(',', '').astype(int)
```

```python
print("2) Average population of the cities where there are more males:")
more_males = df1[df1['Gender Ratio'] > 1]
avg_population = more_males['Population'].mean()
print(f"Average population in cities with more males: {avg_population:,.2f}")
```

# 3)	Places with more males (GenderRatio >1) and less than 2 persons per household:

```python
# 3
df1['Households'] = df1['Households'].astype(str).str.replace(',', '').astype(int)

persons_per_household = df1['Population'] / df1['Households']

filtered_places = df1[
    (df1['Gender Ratio'] > 1) &
    (persons_per_household < 2)
]

print(filtered_places)
```

# 4)	Sort the DataFrame in ascending order of the 'Households.' Show the top 10.

```python
sorted_df1 = df1.sort_values(by='Households', ascending=True)
sorted_df1.head(10)
```
