1) Make a frequency table of 'yearOfStart' and visualize by year.
â€¢	Sort by the year.
â€¢	Draw a line plot from 1997 to 2014. <= Hint: plt.xlim()


```python
# Frequency table of 'yearOfStart' and visualize 
#Frequency table(sorted by year)
year_freq = df3['Year of Start'].value_counts().sort_index()
year_freq
```

```python
#line plot
plt.figure()
plt.plot(year_freq.index, year_freq.values)
plt.xlim(1997, 2014)
plt.xlabel('Year of Start')
plt.ylabel('Number of Coffee Shops')
plt.title('Coffee Shops Started Per Year')
plt.show()
```

2) Split the data by the current state of business ('In' or 'Out' of business). Then, visualize the yearly trend of the 'yearOfStart' frequencies.
â€¢	Sort by the year.
â€¢	Draw two overlapping line plots from 1997 to 2014.
â€¢	Use the 'figure' object.


```python
#split the data
df_in = df3[df3['Current State'] == 'In']
df_out = df3[df3['Current State'] == 'Out']
```

```python
#year frequencies (sorted)
in_freq = df_in['Year of Start'].value_counts().sort_index()
out_freq = df_out['Year of Start'].value_counts().sort_index()
```

```python
#Two overlapping line plots
plt.figure()
plt.plot(in_freq.index, in_freq.values, label='In Business')
plt.plot(out_freq.index, out_freq.values, label='Out of Business')
plt.xlim(1997, 2014)
plt.xlabel('Year of Start')
plt.ylabel('Number of Coffee Shops')
plt.title('Yearly Trend of Coffee Shops (In vs Out of Business)')
plt.legend()
plt.show()
```
