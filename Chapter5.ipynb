```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics, preprocessing

# Load data
data = load_breast_cancer()
X = data['data']
Y = 1 - data['target']  # 1 = malignant, 0 = benign

# Split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=1234)

# Normalize (important for KNN)
scaler = preprocessing.StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# KNN with k=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, Y_train)
Y_pred = knn.predict(X_test_scaled)

# Metrics
accuracy = metrics.accuracy_score(Y_test, Y_pred)
precision = metrics.precision_score(Y_test, Y_pred)  # for class 1 (malignant)
recall = metrics.recall_score(Y_test, Y_pred)        # for class 1

print("Accuracy:", accuracy)
print("Precision (malignant):", precision)
print("Recall (malignant):", recall)
```
