# 3. The EM algorithm belongs to the greedy algorithm. Please explain the reason.

#Answer:- 
The EM (Expectation-Maximization) algorithm is considered a greedy algorithm because:

     1.Local Optimization at Each Step:
In the E-step, it computes expected likelihood based on current parameters. In the M-step, it updates parameters to maximize this expected likelihood locally.
This is a "greedy" choice—it takes the best immediate improvement without considering long-term convergence path.

     2.No Backtracking:
Like typical greedy algorithms, EM never revisits or revises previous parameter estimates. It monotonically increases the likelihood (or evidence lower bound in variational EM) step by step.

     3.Converges to Local Optima:
Because it greedily climbs the likelihood surface, EM often gets stuck in local maxima depending on initialization—a hallmark of greedy optimization.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
import matplotlib.pyplot as plt
```

```python
# Create sample data
X, label = make_blobs(
    n_samples=300,
    n_features=2,
    centers=3,
    cluster_std=5,
    random_state=123
)

# Original data visualization
plt.scatter(X[:, 0], X[:, 1], c=label, alpha=0.7)
plt.title("Dataset #1 : Original")
plt.show()
```

```python
# K-Means Clustering(k=3)
kmeans = KMeans(n_clusters=3, random_state=123)
kmeans_labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, alpha=0.7)
plt.title("K-Means Clustering (k=3)")
plt.show()
```

```python
# Agglomerative Clustering(k=3)
agglo = AgglomerativeClustering(n_clusters=3)
agglo_labels = agglo.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=agglo_labels, alpha=0.7)
plt.title("Agglomerative Clustering (k=3)")
plt.show()
```

```python
# DBSCAN Clutering
dbscan = DBSCAN(eps=5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, alpha=0.7)
plt.title("DBSCAN Clustering")
plt.show()
```
